fillna null value using medium 

```python

df['us_viewers_in_millions']=df['us_viewers_in_millions'].fillna((df['us_viewers_in_millions'].median()))
df['imdb_rating']=df['imdb_rating'].fillna((df['imdb_rating'].median()))
df['views']=df['views'].fillna((df['views'].median()))
df['imdb_votes']=df['imdb_votes'].fillna((df['imdb_votes'].median()))
```
